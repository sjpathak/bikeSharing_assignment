## Model 1: Building a linear model - Manually eliminating variables

#### Steps
<ol>
  <li>Create model using StatsModel OLS.</li>
  <li>Check for p-value and VIF for the variables.</li>
  <li>Keep eliminating variables till p-value is less than 0.05 for remaining variables and VIF is less than 5.</li>
</ol>

```python
# Dividing into X and Y sets for training

y_train = df_train.pop('cnt')
X_train = df_train
```

### Adding all variables to the model and then removing one at a time

```python
# Building a linear model
import statsmodels.api as sm

X_train_lm = sm.add_constant(X_train)

lr_1 = sm.OLS(y_train, X_train_lm).fit()

lr_1.params
```

#### Checking VIF to see which columns to drop - multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Dropping the variable and updating the model

### 4 possibities
- High p-value, High VIF: Remove - variable has low significance (high p-value) and high multicollinearity
- Low p-value, Low VIF: KEEP - Significant variable with low multicollinearity
- High-low
    - High p-value, Low VIF: **remove these first** - Low significance variable - then recreate model, VIF might decrease for other variables
    - Low p-value, High VIF: **remove these after the ones above**
    
#####  NOTE: Remove one column at a time

```python
# given a training set and the columns to remove from the training set, uses statsmodel OLS to create
# a model and return the model.summary and VIF score sorted on above 4 possiblities

# Returns
# 1. variables with p-value > 0.05
# 2. Summary table with R-Squared info
# 3. Training set with columns removed

def train_ols_pvalue_vif(x_train, y_train, column_to_remove = None):
    # Dropping highly correlated variables and insignificant variables
    x_mod = x_train.drop(column_to_remove, axis=1) if column_to_remove is not None else x_train
    
    # Build next fitted model with remaining features
    X_train_lm = sm.add_constant(x_mod)
    lr = sm.OLS(y_train, X_train_lm).fit()
    
    # Extract Features and their corresponding p-value > 0.05, sort descending
    summary_table = lr.summary().tables[1]
    summary = pd.DataFrame(data=summary_table.data[1:], columns=summary_table.data[0])
    summary.rename(columns={summary.columns[0]: 'Features'}, inplace=True)
    summary = summary[['Features', 'P>|t|']].sort_values(by='P>|t|', ascending=False)
    summary['P>|t|'] = summary['P>|t|'].astype(float)
    summary = summary[summary['P>|t|'] > 0.05]
    
    # get VIF for above features
    vif = pd.DataFrame()
    vif['Features'] = X_train_lm.columns
    vif['VIF'] = [variance_inflation_factor(X_train_lm.values, i) for i in range(X_train_lm.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    vif[vif['Features'].isin(summary['Features'].values)]
    
    # returned a merged p-value - VIF table for features
    summary_merged = pd.merge(summary, vif, on='Features', how='inner')
    return summary_merged, lr.summary().tables[0], x_mod
```

### Build next fitted model with remaining features

- Removing: 
    - High p-value, High VIF
    - High p-value, Low VIF

## Model 2: Using RFE (Automatic selection)

```python
# Dividing into X and Y sets for training
# y_train remains unchanged
X_train = df_train
```

### Building model using RFE (Recursive Feature Elimination)

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 20
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, n_features_to_select=20)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
# List the features selected and their ranking

list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
# features (cols) selected by RFE

selected_features = X_train.columns[rfe.support_]
selected_features
```

```python
# cols not selected by RFE

X_train.columns[~rfe.support_]
```

```python
# scale train and test data

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Fit the model using only the selected features
rfe_model = LinearRegression()
rfe_model.fit(X_train_scaled[:, rfe.support_], y_train)
```

```python
# Make predictions on the training data using the fitted RFE model
y_train_pred = rfe.predict(X_train)

# number of observations
n = X_train.shape[0]
# number of selected features
p = rfe.n_features_

# Calculate R-squared and MSE on the training data
r2_train = r2_score(y_train, y_train_pred)
mse_train = mean_squared_error(y_train, y_train_pred)
adjusted_r2_train = 1 - (1 - r2_train) * (n - 1) / (n - p - 1)

print(f'Training Data Results:')
print(f'\tR-squared: {r2_train}')
print(f'\tAdjusted R-squared on training data: {adjusted_r2_train}')
print(f'\tMean Squared Error on training data: {mse_train}')
```

### RFE Model fit test on test data

```python
# Make predictions on the test data
y_pred = rfe_model.predict(X_test_scaled[:, rfe.support_])
```

```python
# Evaluate the model
r2_test = r2_score(y_test, y_pred)
mse_test = mean_squared_error(y_test, y_pred)
adjusted_r2_test = 1 - (1 - r2) * (len(y_test) - 1) / (len(y_test) - X_test_scaled[:, rfe.support_].shape[1] - 1)

print(f'Test Data Results:')
print(f'\tR-squared: {r2_test}')
print(f'\tAdjusted R-squared on training data: {adjusted_r2_test}')
print(f'\tMean Squared Error on training data: {mse_test}')
```
